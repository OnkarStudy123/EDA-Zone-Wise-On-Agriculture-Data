```python
df=data[['Year','State Name','Dist Name','RICE PRODUCTION (1000 tons)','WHEAT PRODUCTION (1000 tons)','CHICKPEA PRODUCTION (1000 tons)','GROUNDNUT PRODUCTION (1000 tons)','SUNFLOWER PRODUCTION (1000 tons)','SOYABEAN PRODUCTION (1000 tons)','SUGARCANE PRODUCTION (1000 tons)','COTTON PRODUCTION (1000 tons)']]
```

```python
df.nunique()
```

# West Zone Devide

```python
west_zone = df[df['State Name'].isin(['Maharashtra', 'Madhya Pradesh', 'Gujarat', 'Chhattisgarh', 'Rajasthan'])]
west_zone
```

# South Zone Devide

```python
south_zone=df[df['State Name'].isin(['Andhra Pradesh','Karnataka','Kerala','Tamil Nadu','Telangana'])]
south_zone
```

# North Zone Devide

```python
north_zone=df[df['State Name'].isin(['Haryana','Himachal Pradesh','Punjab','Uttar Pradesh','Uttarakhand'])]
north_zone
```

# East Zone Divide

```python
east_zone=df[df['State Name'].isin(['Assam','Bihar','Jharkhand','West Bengal'])]
east_zone
```

## In below code we will concat the two zones in one data

```python
new_data_1 = pd.concat([north_zone, east_zone], ignore_index=True)
new_data_1
```

## In below code we will concat the two zones in one data

```python
new_data_2 = pd.concat([west_zone,south_zone],ignore_index=True)
new_data_2
```

```python
filter_data_ne = new_data_1[(new_data_1 !=0).all(axis=1)]
filter_data_ne
```

```python
filter_data_ws = new_data_2[(new_data_2 !=0).all(axis=1)]
filter_data_ws
```

### north east zone Rice Production 

```python
# Aggregate production by state
production_by_state = filter_data_ne.groupby('State Name').sum()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(production_by_state['RICE PRODUCTION (1000 tons)'], labels=production_by_state.index, autopct='%1.1f%%', startangle=140)
plt.title('Rice Production Distribution by north and east zones')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### West and South zone Wheat Production

```python
# Aggregate production by state
production_by_state = filter_data_ws.groupby('State Name').sum()

# Create a pie chart
plt.figure(figsize=(6,6))
plt.pie(production_by_state['WHEAT PRODUCTION (1000 tons)'], labels=production_by_state.index, autopct='%1.1f%%', startangle=140)
plt.title('wheat Production Distribution by west and south zones')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### North And East Zone  Wise Wheat Production

```python
plt.figure(figsize=(8, 6))
filter_data_ne.groupby('State Name')['WHEAT PRODUCTION (1000 tons)'].sum().plot(kind='bar', color='skyblue')
plt.ylabel('Total Production')
plt.xticks(rotation=45)
plt.show()
```

### North and East zone wise Chikpea Production 

```python
plt.figure(figsize=(8, 6))
filter_data_ne.groupby('State Name')['CHICKPEA PRODUCTION (1000 tons)'].sum().plot(kind='bar', color='skyblue')
plt.ylabel('Total Production')
plt.xticks(rotation=45)
plt.show()

```

### North And East Zone Wise Groundnut Production

```python
plt.figure(figsize=(6, 6))
filter_data_ne.groupby('State Name')['GROUNDNUT PRODUCTION (1000 tons)'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Production by Crop')
plt.xlabel('Crop')
plt.ylabel('Total Production')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot to explore relationship between groundnut production and state name
plt.figure(figsize=(10, 6))
plt.scatter(filter_data_ne['State Name'], filter_data_ne['GROUNDNUT PRODUCTION (1000 tons)'])
plt.title('Groundnut Production by State')
plt.xlabel('State')
plt.ylabel('Groundnut Production (1000 tons)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
